Atleta con el menor número de apariciones y mayores ganancias

```python
#Para eliminar las apariciones repetidas de atletas 
x = df.drop_duplicates(['Name'])
#Para obtener la información correspondiente al deportista que aparece menos y gana más
y = x.loc[x['earnings ($ million)'] == x['earnings ($ million)'].max()]
#Ganacia de dicho deportista
Earning = str(y.iloc[0]['earnings ($ million)'])
#Deportista
Athlete = y.iloc[0]['Name']
#Imprimimos la información pedida
print('El deportista con menos apariciones en el dataset y mayores ganacias es', Athlete,
      'con una ganacia de $' + Earning, 'millones de dólares.')
```

Deporte y país con mayor número de atletas no rankeados que entraron en la lista de atletas mejor pagados

```python
df['Sport'] = df ['Sport'].str.capitalize()
df = df.where(df != '?', 'NaN')
df = df.where(df != '??', 'NaN')
df = df.where(df != 'not ranked', 'NaN')
df = df.where(df != 'none', 'NaN')
notranked = pd.DataFrame(df[df['Previous Year Rank'].isna()])
X = notranked.groupby('Sport').count()
Y = notranked.groupby('Nationality').count()
Deporte = X.index[(X['Name'] == X['Name'].max())].tolist()
Deporte[0]
Pais = Y.index[(Y['Name'] == Y['Name'].max())].tolist()
Pais[0]
print('El deporte con mayor número de atletas no rankeados que entraron en la lista de atletas mejor pagados es', Deporte[0])
print('y el país con mayor número de atletas no rankeados que entraron en la lista de atletas mejor pagados es', Pais[0])
```


País con mayor número de deportes con atletas en el dataset

```python
#Primero estandarizamos el formato de los datos de la columna 'Sport' ya que difieren en formato las observaciones
df['Sport'] = df ['Sport'].str.capitalize()
#Creamos una nueva columna de nacionalidad y deporte
df['Nationality-Sport'] = df['Nationality'] + ' - ' + df['Sport']
#Para así conservar sólo una observación asociada a deporte por cada país
z = df.drop_duplicates(['Nationality-Sport'])
#Eliminamos la columna auxiliar, devolviendo el dataframe a su estado inicial
df = df.drop('Nationality-Sport', axis = 1)
#Contamos cuántos deportes hay por cada país
w = z.groupby(z['Nationality']).count()
#Tomamos el país con mayor número de deportes 
Country = w.index[(w['Sport'] == w['Sport'].max())].tolist()
Country[0]
#Imprimimos la información pedida
print('El país que cuenta con mayor número de deportes en el dataset es',Country[0],
      'con', w['Sport'].max(), 'deportes distintos.')
```

Ganancia mínima y máxima (dentro del dataset) por deporte y por país.

```python
#Primero estandarizamos el formato de los datos de la columna 'Sport' ya que difieren en formato las observaciones
df['Sport'] = df['Sport'].str.capitalize()
#Creamos una nueva columna de nacionalidad y deporte 
df['Nationality-Sport'] = df['Nationality'] + '-' + df['Sport']
#Fijamos las columnas que nos interesan 
atle = df.loc[:,('Nationality','Sport', 'earnings ($ million)')]
#Obtenemos las ganacias mínimas por deporte por país
min = atle.groupby(['Sport', 'Nationality'])['earnings ($ million)'].min()
#Obtenemos las ganacias máximas por deporte por país
max = atle.groupby(['Sport', 'Nationality'])['earnings ($ million)'].max()
#Las convertimos en listas para crear un nuevo dataframe
Min = list(min)
Max = list(max)
#Eliminamos datos repetidos de deporte y país
x = df.drop_duplicates(['Nationality-Sport'])
#Ordemanos los datos por deporte por país para se ajuste a las listas 
y = x.sort_values(by = ['Sport', 'Nationality']) 
#Creamos el nuevo dataframe con lo desarrollado anteriormente
MinMax = pd.DataFrame({'Sport' : y['Sport'], 'Country' : y['Nationality'], 'Min' : Min, 'Max' : Max})
MinMax
```

```python
#Separamos las columnas con las que vamos a trabajar
df = df.loc[:,('Nationality', 'Sport', 'earnings ($ million)')]
#Para pasar los datos con la primer letra mayúscula y los demás en minúscula
df['Sport'] = df['Sport'].str.capitalize()
#df
#Creamos un grupo para agrupar los valores
min= df.groupby(['Nationality','Sport'])['earnings ($ million)'].min()
#min
#Para convertir una lista a un DataFrame
mindf = pd.DataFrame(min)
#mindf
#Para ordenar de manera ascendente
minimo = mindf.sort_values(by =['earnings ($ million)'], axis=0, ascending=True)

#Separamos las columnas con las que vamos a trabajar
df = df.loc[:,('Nationality', 'Sport', 'earnings ($ million)')]
#Para pasar los datos con la primer letra mayúscula y los demás en minúscula
df['Sport'] = df['Sport'].str.capitalize()

#Creamos una lista con los valores agrupados
max= df.groupby(['Nationality','Sport'])['earnings ($ million)'].max()

#Para convertir una lista a un DataFrame
maxdf = pd.DataFrame(max)

#Para ordenar de manera ascendente
maximo = maxdf.sort_values(by =['earnings ($ million)'], axis=0, ascending=True)
print('La ganancia minima es de: ', minimo['earnings ($ million)'].min(), minimo.index[0],'y la ganancia máxima es de ',maximo['earnings ($ million)'].max(), maximo.index[0])
```
